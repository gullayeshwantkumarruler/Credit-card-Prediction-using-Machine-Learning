```python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('Salaries.csv')
df.head()
```

## 1.Bar-Plot

```python
# 1. Bar-Graph
# 1) Show the frequencies of each level of ‘rank’ column.

df['rank'].value_counts()
```

```python
# 2) Draw the bar-graph of the column 'rank' with any single color.
# Make the borderline for bars and make the figure size (4,4) .

plt.figure(figsize=(4,4))
plt.bar(df['rank'].value_counts().index, df['rank'].value_counts().values, color='red', edgecolor='black')
plt.show()
```

```python
# 3) Refer to the RGB color table (https://www.rapidtables.com/web/color/RGB_Color.html).
# Use ‘#99FF33’ for ‘Prof’, ‘#33F9FF’ for AssocProf, and ‘#F9FF33’ for AsstProf to draw new bar
# graph that has different colors of each bar. Use ‘blue’ color for the edge of the bar.

plt.figure(figsize=(4,4))
plt.bar(df['rank'].value_counts().index, df['rank'].value_counts().values, color=['#99FF33', '#33F9FF', '#F9FF33'], edgecolor='blue')
plt.show()
```

```python
# 4) Add title “Frequency dist. of professors by rank”, x-axis label “Rank”, and y- axis label
# “Frequency” to the bar graph. Add frequency of each rank on the top of each bar as follows.
# To display the frequency of each class on the bar-graph, refer to the websites:
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.text.html
# https://www.geeksforgeeks.org/display-percentage-above-bar-chart-in-matplotlib/
# https://stackoverflow.com/questions/53066633/how-to-show-values-on-top-of-bar-plot

plt.figure(figsize=(4,4))
plt.bar(df['rank'].value_counts().index, df['rank'].value_counts().values, color=['#99FF33', '#33F9FF', '#F9FF33'], edgecolor='blue')
plt.title('Frequency dist. of professors by rank')
plt.xlabel('Rank')
plt.ylabel('Frequency')
plt.text(0, 100, '100', ha='center', va='bottom')
plt.text(1, 50, '50', ha='center', va='bottom')
plt.text(2, 50, '50', ha='center', va='bottom')
plt.show()

```

## Box-plot and histogram

```python

# 1) Display the 5-numbers( minimum, 25%, 50%, 75% percentiles and maximum) of ‘phd’ column.
df['phd'].describe()
```

```python

# 2) Check if there are missing values of ‘phd’ column.
df['phd'].isnull().sum()
```

```python

# 3) Draw the box-plot of ‘phd’ column.
df['phd'].plot.box()
```

```python

# 4) Use 10-point width to construct the classes of ‘phd’ column as follows:
# [0, 10), [10, 20), ... , [50, 60), respectively. Then, find the frequency of each class and display
# it. That is, find the number of observations that belong to each class.
df['phd'].value_counts(bins=[0,10,20,30,40,50,60],sort=False)
```

```python

# 5) Draw the histogram of 'phd' column with the frequency distribution obtained from (4) as
# follows. Use yellow-colored bar and blue borderline for each bar.
df['phd'].plot.hist(bins=6, color='yellow', edgecolor='blue')
```

```python

# 6) Draw the bar-graph of 'phd' column with different colors of each bar, appropriate title, axes
# labels and frequency on the top of each bar as follows. For the color of each bar in the bargraph,
# select 6 different colors from the RGB color table
# (https://www.rapidtables.com/web/color/RGB_Color.html).
df['phd'].value_counts(bins=[0,10,20,30,40,50,60],sort=False).plot.bar(color=['red', 'green', 'blue', 'yellow', 'orange', 'purple'])
plt.title('Histogram of years since PHD')
plt.xlabel('Years since PHD')
plt.ylabel('Frequency of PhDs')
plt.show()

```

```python

# 7) State the findings and results from the box-plot and histogram, respectively.
# The box-plot shows that the median is 20, the first quartile is 15, and the third quartile is 25.
# The histogram shows that the most common number of years of education is 20.
```

## GroupBy plot

```python

# 1) Construct frequency cross-tabulation for ‘rank’ and ‘sex’
df1 = pd.crosstab(df['rank'],df['sex'])
df1
```

```python

# 2) Draw the pie chart of the frequency distribution of ‘rank’ and ‘sex’ that is obtained from (1).
# You should use Yellow color for Male and Blue color for Female group.
df1.plot.pie(subplots=True, colors=['yellow','blue'], autopct='%.2f%%', figsize=(10,5))
```

```python

# 3) Calculate the mean salary of each level of ‘rank’ and ‘sex’ column. Use 2 decimal places.
# That is , there are six groups of ‘rank’ and ‘sex’ column : (Prof., Female), (Prof., Male), (AssocProf., Female), (AssocProf., Male), (AsstProf., Female), and (AsstProf., Male).
df2 = df.groupby(['rank','sex']).mean().round(2)
df2
```

```python

# 4) Draw the bar graph of the mean salary of each level of 'rank' and ‘sex’.
# Select any color but use different color for each bar.
df2['salary'].plot.bar(color=['red','blue','green','yellow','black','pink'], figsize=(10,5))
```

```python
 # 5) Plot the stacked bar graph of the mean salary of each level of 'sex' and ‘rank’.
 # This can be done in two ways
 # first method
df2['salary'].unstack().plot.bar(stacked=True, figsize=(10,5))
```

```python
# second method
df2['salary'].unstack().plot.bar(figsize=(10,5))
```

```python
# 6) State the findings and results from the box-plot and histogram, respectively.
df['salary'].plot.box(figsize=(10,5))
```
